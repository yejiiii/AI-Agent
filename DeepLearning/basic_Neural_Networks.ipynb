```python
import tensorflow as tf
import numpy as np


#############################################################
#dataset
# sex, age, income
X_train = np.array(([1.0, 0.7, 0.17], #1
                    [-1.0, -1, -0.6], #2
                    [-1.0, 0.25, 0.4], #3
                    [1.0, -0.1, 0.6], #4
                    [-1.0, -0.75, -1], #5
                    [1.0, -0.5, 0.27], #6
                    [1.0, 0.4, -0.27], #7
                    [-1.0, -0.4, -0.33], #8
                   ), dtype=float)

# vote(+), vote(-)
y_train = np.array(([1, 0], #1
                    [0, 1], #2
                    [1, 0], #3
                    [0, 1], #4
                    [1, 0], #5
                    [0, 1], #6
                    [1, 0], #7
                    [1, 0], #8
                   ), dtype=float)

X_test = np.array(([[1.0, -0.55, -0.23]]), dtype=float)
y_test = np.array(([[0, 1]]), dtype=float)
#############################################################




#############################################################
# Define NN model
X = tf.placeholder(tf.float32) #input
Y = tf.placeholder(tf.float32) #output

# w1 [input -> hidden] = [3, 10]
w1 = tf.Variable(tf.random_uniform([3, 10], -1., 1.))
# w1 [hidden -> output] = [10, 2]
w2 = tf.Variable(tf.random_uniform([10, 2], -1., 1.))

# bias: the number of output neuron at each layer
b1 = tf.Variable(tf.zeros([10]))
b2 = tf.Variable(tf.zeros([2]))
#############################################################





#############################################################
# Build a graph

# hidden from input
layer1_weighted_sum = tf.add(tf.matmul(X, w1), b1) #weighted sum
layer1_output = tf.nn.sigmoid(layer1_weighted_sum) #activation function

# output from hidden
logits = tf.add(tf.matmul(layer1_output, w2), b2)

# LOSS(COST): cross entropy 
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=Y, logits=logits))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train_op = optimizer.minimize(cost)
#############################################################




#############################################################
#train
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for step in range(10000):
    sess.run(train_op, feed_dict={X: X_train, Y: y_train})
    if (step + 1) % 1000 == 0:
        print("epoch: " + str(step + 1))
        print("loss: " + str(sess.run(cost, feed_dict={X: X_train, Y: y_train})))
        #training validation
        prediction = tf.argmax(logits, 1)
        target = tf.argmax(Y, 1)
        print('train actual', sess.run(target, feed_dict={Y: y_train}))
        print('train pred:', sess.run(prediction, feed_dict={X: X_train}))
        print()
#############################################################




#############################################################
#evaluation
prediction = tf.argmax(logits, 1)
target = tf.argmax(Y, 1)
print("#################################")
print("results")
print('test actual', sess.run(target, feed_dict={Y: y_test}))
print('test pred:', sess.run(prediction, feed_dict={X: X_test}))

is_correct = tf.equal(prediction, target)
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
print('accuracy: %.2f' % sess.run(accuracy * 100, feed_dict={X: X_test, Y: y_test}))
```
